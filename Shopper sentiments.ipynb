```python
import pandas as pd
import warnings

warnings.filterwarnings('ignore',category=FutureWarning)
```

```python
df = pd.read_csv(r"C:\Users\isaac\Downloads\TeePublic_review.csv",encoding='latin-1')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.hist(df['review-label'])
plt.title('Distribution of Ratings')
plt.xlabel('Review Labels')
plt.ylabel('Frequency')
plt.show()
```

```python
grouped_df = df.groupby(['year', 'month']).size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.barplot(data=grouped_df, x='month', y='count', hue='year')

plt.xlabel('Month')
plt.ylabel('Review Count')
plt.title('Monthly Review Count Over the Years')
plt.legend(title='Year')

plt.show()
```

```python
plt.scatter(df['longitude'], df['latitude'], c=df['review-label'], cmap='viridis', s=10)
plt.colorbar(label='Review Label')
plt.title('Geospatial Distribution of Reviews')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

```python
plt.boxplot([df[df['year'] == year]['review-label'] for year in df['year'].unique()], labels=df['year'].unique())

plt.xlabel('Year')
plt.ylabel('Review Label')
plt.title('Distribution of Ratings by Year')

plt.show()
```
